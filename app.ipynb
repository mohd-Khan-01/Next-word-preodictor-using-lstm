```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM, Embedding, Dropout
```

```python
import pandas as pd
df = pd.read_table(
    "/root/.cache/kagglehub/datasets/muhammadbilalhaneef/sherlock-holmes-next-word-prediction-corpus/versions/1/Sherlock Holmes.txt",
    header=None,
    names=["text"],
    encoding="utf-8"
)
```

```python
df.head(10)
```

```python
df['text'] = df['text'].str.strip()
```

```python
#Remove empty lines
df = df[df["text"] != ""].reset_index(drop=True)
```

```python
texts = df['text'].drop_duplicates().astype(str).tolist()
```

```python
texts
```

```python
document = '\n'.join(texts).lower()
```

```python
document
```

```python
sentences = document.split('\n')
```

```python
sentences
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
token=Tokenizer(oov_token='<unkn>')
```

```python
token.fit_on_texts(sentences)
```

```python
token.word_index
```

```python
token.word_counts
```

```python
sequences = token.texts_to_sequences(sentences)
```

```python
sequences
```

```python
training_seq=[]
for seq in sequences:
  for i in range (1,len(seq)):
    training_seq.append(seq[:i+1])
```

```python
training_seq[:30]
```

```python
max_len=max(len(seq) for seq in sequences)
```

```python
max_len
```

```python

padding_sequences=pad_sequences(training_seq,maxlen=max_len,padding="pre")
```

```python
X=padding_sequences[:, :-1]
y=padding_sequences[:,-1]
```

```python
vocab_size = len(token.word_index) + 1
```

```python
model=Sequential([
   Embedding(input_dim=vocab_size,output_dim=100),
   LSTM(150),
   Dense(vocab_size,activation="softmax")
])
```

```python
model.build(input_shape=(None,max_len))
model.compile(loss = 'sparse_categorical_crossentropy', optimizer = 'adam', metrics = ['accuracy'])

model.summary()
```

```python
history=model.fit(X,y,epochs=50)
```

```python
import numpy as np
def next_word_predict(model,text,tokenizer,max_len):
  token_seq=tokenizer.texts_to_sequences([text])
  padded_seq = pad_sequences(
        token_seq,
        maxlen=max_len - 1,
        padding='pre'
    )
  prob=model.predict(padded_seq,verbose=0)
  index=np.argmax(prob)
  next_word="<unkn>"
  for word, idx in tokenizer.word_index.items():
    if idx == index:
      next_word = word
      break

  text = text + " " + next_word
  return text
```
